```python
import os
import pandas as pd
from goatools.gosubdag.gosubdag import GoSubDag
import csv

from goatools.obo_parser import GODag
```

```python
godag = GODag('../tests/data/i126/go_2021.obo',
              optional_attrs={'relationship'})
```

```python
filename='GotermList.tab'
df = pd.read_csv(filename,sep='\t')
All_go_term_list=df['Goterm'].tolist()
```

```python
output_file='Goterm_Depth.csv'
with open(output_file,'w') as outputcsv_file:
    spamwriter = csv.writer(outputcsv_file,delimiter=',')
    col1=[]
    col1.append("GO ID")
    col1.append("Depth")
    spamwriter.writerow(col1)
    for i in range(0,len(All_go_term_list)):
        col1=[]
        go_id=All_go_term_list[i]
        flag=1
        try:
            go_term=godag[go_id]
            Depth = go_term.depth
            col1.append(go_id)
            col1.append(Depth)
            print(col1)
        except:
            flag=0
        if flag==1:
            spamwriter.writerow(col1)
        
```
